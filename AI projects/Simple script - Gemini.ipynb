## **Using the Requests library**

```python
# Calling the Gemini API with a script
import requests
import json


# Get the API key from the user data in collab
API_TOKEN = ('AIzaSyBxXQYrp4RMA9QB9uDjVaO586iQ7lw4bzI')

# Base Url of Gemini
base_url_gemini_api = "https://generativelanguage.googleapis.com/v1beta/models/gemini-2.5-flash:generateContent"

# Headers of the content type
headers = {"Content-Type": "application/json"}

# The prompt in json format, which was defined like that before
prompt = {
  "contents": [
    {
      "parts": [
        {
          "text": "Explicame que es la inteligencia artificial"
        }
      ]
    }
  ]
}


# Send the prompt to Gemini and the parameters given
response = requests.post(base_url_gemini_api, headers=headers, params = {'key': API_TOKEN}, json = prompt)


# Extracting Gemini's response

if response.status_code == 200:
  response_data = response.json()
  gemini_response = response_data['candidates'][0]['content']['parts'][0]['text']
  print(gemini_response)

else:
  print(f"Error en: {response.status_code}")
```
